```python
from IPython.display import clear_output
import time

# all_properties = {# rentalproperty.name: [annual_cashflow, total_investment]}
all_properties = {}
# all_properties = {'123 Appaloosa Run': [2820, 47000], '456 Whispering Wind Dr.': [600, 28000], '1800 Real Street': [18120, 33000]}

def to_float(string):
    price = string
    is_float = False
    while is_float == False:
        try:
            float(price)
            return float(price)
        except ValueError:
            price = input("Invalid response. Please enter a number.").strip('$ ')
            
def to_int(string):
    if string.isdigit():
        return int(string)
    else:
        while string.isdigit() == False:
            string = input("Invalid response. Please enter a number.").strip()
        return int(string)
    
def big_ROI(dict1):       # if someone has multiple rental properties and wants their overall ROI for all properties put together
    cash_sums = 0
    invest_sums = 0
    # add up all annual_cashflows
    for x in all_properties.values():
        cash_sums += x[0]
    # add up all total_investments
    for x in all_properties.values():
        invest_sums += x[1]
    # (annual_cashflows / total_investments) * 100
    total_percentage = (cash_sums/invest_sums) * 100
    print('Your properties:')
    print('\t--------------------------------')
    for key, value in all_properties.items():
        print(f"\t{key}: {(value[0]/value[1])*100:.2f}%")
    print('\t--------------------------------')
    print(f"\nYour cash on cash ROI for all of your properties combined is {total_percentage:.2f}%.")
    return f"{total_percentage:.2f}%"  

```

```python
class Landlord:
    def __init__(self, name):
        self.name = name
        
    def __str__(self):
        return f"{self.name}"
    
    def calculate(self):
        greeting = input(f"Hi {self.name}! Let's figure out your Cash-on-Cash Return. Press return to get started. ")
        clear_output()
        start = True
        while start == True:
            prop_name = input("Enter the name of the property that we will calculate the ROI for: ")
            property1 = RentalProperty(prop_name)
            property1.income()
            property1.expenses()
            property1.cash_flow()
            move_on = input("Press return when ready to move on. ")
            clear_output()
            property1.ROI()
            if len(all_properties) == 1:
                another = input('Would you like to find the ROI on another property? y/n ')
                while another not in ('y', 'n'):
                    another = input('Invalid response. Please enter y or n. ')
                if another == 'n':
                    print('Thank you for using our ROI calculator!')
                    return
                else:
                    start == True
            elif len(all_properties) > 1:
                another = input('Would you like to find the ROI on another property? y/n ')
                while another not in ('y', 'n'):
                    another = input('Invalid response. Please enter y or n. ')
                if another == 'n':
                    see_all = input('Would you like to calculate the combined ROI for all of your properties? y/n ')
                    clear_output()
                    while see_all not in ('y', 'n'):
                        see_all = input('Invalid response. Please enter y or n. ')
                    if see_all == 'n':
                        print('Thank you for using our ROI calculator!')
                        return
                    else: 
                        big_ROI(all_properties)
                        print('\nThank you for using our ROI calculator!')
                        return
                else:
                    start == True
                
                
        
class RentalProperty:
    def __init__(self, prop_name):       
        self.prop_name = prop_name
        self.total_income = 0
        self.total_expenses = 0
        self.annual_cash_flow = 0
        self.total_investment = 0
        self.return_on_investment = 0
        
    def __str__(self):
        return f"{self.prop_name}"
    
    def __repr__(self):
        return f"<Property|{self.prop_name}>"
          
    def income(self):
        print(f"First, let's calculate your monthly income for {self.prop_name}.")
        time.sleep(2)
        units = to_int(input(f'How many units does {self.prop_name} have? '))
        rent_charge = to_float(input('How much rent do you charge for each unit per month? '))
        rental_income = units * rent_charge
        extra_income = input('Do you have any other sources of income for this property? y/n ')
        clear_output()
        while extra_income not in ('y', 'n'):
            extra_income = input('Invalid response. Please enter y or n. ')
        if extra_income == 'y':
            laundry = to_int(input('How much laundry income do you receive each month? Enter 0 if none. '))
            storage = to_int(input('How much storage income do you receive each month? Enter 0 if none. '))
            miscellaneous = to_int(input('How much income do you receive from miscellaneous sources each month? Enter 0 if none. '))
            clear_output()
            self.total_income = rental_income + laundry + storage + miscellaneous
            print(f"Your total monthly income for {self.prop_name} is ${self.total_income:.2f}.\n")
            return self.total_income
        else:
            self.total_income = rental_income
            print(f"Your total monthly income for {self.prop_name} is ${self.total_income:.2f}.\n")
            return self.total_income
    
    def expenses(self):
        # return total monthly expenses as float
        print(f"Next let's determine your monthly expenses for {self.prop_name}.")
        print("For each expense, please enter how much you spend per month. Enter 0 if not applicable.")
        time.sleep(2)
        taxes = to_int(input("\tTaxes: "))
        insurance = to_int(input("\tInsurance: "))
        hoa = to_int(input("\tHOA fee: "))
        lawncare_snow = to_int(input("\tLawncare/snow removal: "))
        vacancy = to_int(input("\tVacancy: "))
        repairs = to_int(input("\tRepairs: "))
        cap_ex = to_int(input("\tCapital Expenditures: "))
        property_mgmt = to_int(input("\tProperty Management: "))
        mortgage = to_int(input("\tMortgage: "))
        utilities = input("Do you pay for any utilities, such as electric/water/sewer/trash/gas? y/n")
        while utilities not in ('y', 'n'):
            utilities = input('Invalid response. Please enter y or n. ')
        if utilities == 'n':
            clear_output()
            self.total_expenses = taxes + insurance + hoa + lawncare_snow + vacancy + repairs + cap_ex + property_mgmt + mortgage
            print(f"Your total monthly income for {self.prop_name} is ${self.total_income:.2f}.\n")
            print(f"Your total monthly expenses for {self.prop_name} are ${self.total_expenses:.2f}.\n")
            return self.total_expenses
        else:
            electric = int(input("\tElectric: "))
            water = int(input("\tWater: "))
            sewer = int(input("\tSewer: "))
            trash = int(input("\tTrash: "))
            gas = int(input("\tGas: "))
            clear_output()
            self.total_expenses = taxes + insurance + hoa + lawncare_snow + vacancy + repairs + cap_ex + property_mgmt + mortgage + electric + water + sewer + trash + gas
            print(f"Your total monthly income for {self.prop_name} is ${self.total_income:.2f}.\n")
            print(f"Your total monthly expenses for {self.prop_name} are ${self.total_expenses:.2f}.\n")
            return self.total_expenses
    
    def cash_flow(self):
        # income - expenses = monthly cashflow
        # monthly cashflow * 12 = annual cashflow
        # return monthly cashflow and annual cashflow as floats
        print("Our next step is to calculate your monthly cash flow, which is the monthly income minus monthly expenses.\n")
        time.sleep(2)
        monthly_cash_flow = self.total_income - self.total_expenses
        print(f"\t{self.total_income:.2f} - {self.total_expenses:.2f} = {monthly_cash_flow:.2f}\n")
        self.annual_cash_flow = monthly_cash_flow * 12
        print(f"Your monthly cash flow is ${monthly_cash_flow:.2f}.\n")
        print(f"Your annual cash flow for {self.prop_name} is ${self.annual_cash_flow:.2f}.")
        return self.annual_cash_flow
    
    def ROI(self):
        # determine total investment
        # add this instance to all_properties dict
        # (annual cashflow / total investment) * 100
        # return percentage for one property instance
        print(f"Our last step is to determine your total investment for {self.prop_name}.\n")
        time.sleep(2)
        down_payment = to_int(input("What did you put down for the down payment? "))
        closing_costs = to_int(input("What were your closing costs? "))
        rehab = to_int(input("Did you spend any on repairs? "))
        misc = to_int(input("Please enter any miscellaneous costs: "))
        clear_output()
        self.total_investment = down_payment + closing_costs + rehab + misc
        print(f"Your total investment for {self.prop_name} is ${self.total_investment:.2f}.\n")
        all_properties[self.prop_name] = [self.annual_cash_flow, self.total_investment]
        self.return_on_investment = (self.annual_cash_flow/self.total_investment)*100
        print('Calculating...\n')
        time.sleep(2)
        print(f"And finally, your cash-on-cash ROI for {self.prop_name} is {self.return_on_investment:.2f}%.")
        return self.return_on_investment
```
